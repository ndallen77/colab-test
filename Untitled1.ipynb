```
import cartopy.mpl.ticker as cticker

def plotMap2():
    proj = ccrs.Mercator(central_longitude=180, min_latitude=15, 
    max_latitude=55)

    fig, ax = plt.subplots(subplot_kw=dict(projection=proj), figsize=(8,8))

    ax.set_extent([255 ,115, 0, 60], crs=ccrs.PlateCarree())

    ax.add_feature(cfeat.LAND, facecolor=[0.0, 0.5, 0.0])
    ax.add_feature(cfeat.LAKES, alpha=0.9)  
    ax.add_feature(cfeat.BORDERS, zorder=10)
    ax.add_feature(cfeat.COASTLINE, zorder=10)

    states_provinces = cfeat.NaturalEarthFeature(
            category='cultural',  name='admin_1_states_provinces_lines',
            scale='50m', facecolor='none')
    ax.add_feature(states_provinces, edgecolor='black', zorder=10)

    
    latMax = 60
    latMin = 0
    lonMax = -120
    lonMin = 120

    xTicks = np.arange(lonMin, lonMax+1, 30)
    yTicks = np.arange(latMin, latMax+1, 10)
    
    ax.set_xticks([120., 140., 160., 180., -160., -140., -120.], crs=ccrs.PlateCarree())
    ax.set_xticklabels([120., 140., 160., 180., -160., -140., -120.], color='red', weight='bold')
    
#    ax.set_xticks(xTicks, crs=ccrs.PlateCarree())
#    ax.set_xticklabels(xTicks, color='red', weight='bold')
    
    ax.set_yticks(yTicks, crs=ccrs.PlateCarree())
    ax.set_yticklabels(yTicks)
    ax.yaxis.tick_right()

    lon_formatter = cticker.LongitudeFormatter()
    lat_formatter = cticker.LatitudeFormatter()
    ax.xaxis.set_major_formatter(lon_formatter)
    ax.yaxis.set_major_formatter(lat_formatter)
    ax.grid(linewidth=2, color='black', alpha=0.5, linestyle='--')
    
plotMap2()
```
